```python
import pandas as pd
```

```python
sales = pd.read_csv('s3://data-analysis-and-visualization-lab/car_data.csv')
```

```python
count_by_month = sales.groupby('sold_month').count()[['car']]
count_by_month.plot.bar()
```

```python
total_price_by_month = sales.groupby('sold_month').sum()[['price']]
total_price_by_month.plot.bar()
```

```python
count_by_salesman = sales.groupby('salesman').count()[['car']]
count_by_salesman.plot.bar()
```

```python
count_by_dealership = sales.groupby('dealership').count()[['car']]
count_by_dealership.plot.bar()
```

```python
sales['sold_year'] = pd.to_datetime(sales['sold_date']).dt.year
```

```python
mean_corvette_price_by_year = sales[sales.car == 'Corvette'].groupby('sold_year').mean()[['price']]
mean_corvette_price_by_year.plot.bar()
```
